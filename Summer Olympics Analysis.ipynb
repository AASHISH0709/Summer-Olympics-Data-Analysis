# 7. Which event is most popular in terms on number of players? (Top 5)

```python
import pandas as pd
import numpy as np


df = pd.read_csv("summer.csv")

data = []

for Event in df['Event'].unique():
    data.append([Event , len(df[df['Event']  == Event])])

pd.DataFrame(data,columns = ['Event','Players']).sort_values(by='Players', ascending=False).head().plot(x = 'Event', y = 'Players', kind = 'bar', figsize = (15,5))
```
